```python
#import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df= pd.read_csv("/content/diabetes_prediction_dataset.csv")
```

```python
#relplot- Relationship Plot

sns.relplot(x="gender", y="diabetes", data=df)
plt.show()
#it is similar to the scatter plot in matplotlib
```

```python
#lineplot -trend analysis
sns.lineplot(data=df, x="gender", y="diabetes")
plt.xticks(rotation=45)
plt.show()
```

```python
#barplot- trend analysis
sns.barplot(data=df, x="age", y="diabetes")
plt.show()
```

```python
#countplot- count categories
sns.countplot(data=df, x="gender")
plt.show()
```

```python
#boxplot- outlier detection
sns.boxplot(data=df, x="gender", y="age")
plt.show()
```

```python
#violinplot - Distribution shape
sns.violinplot(data=df, x="gender", y="diabetes")
plt.show()
```

```python
#histplot- Distribution
sns.histplot(data=df, x="age" , bins =30 , kde= True)
plt.show()
```

```python
#heatmap - correlation matrix
corr= df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
#pairplot()- all variable relationship
sns.pairplot(df[["age","gender","diabetes"]])
plt.show()
```
